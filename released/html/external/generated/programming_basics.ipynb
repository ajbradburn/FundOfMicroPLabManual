```python
import os

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import schemdraw as schem
from schemdraw import flow as f
import schemdraw.elements as e

schem.use('svg')
```

```python
img = schem.Drawing()

img += (start := f.Start().label('Hunger').at((4,3)))
img += (loaf := f.Process(w=4).label('Loaf of Bread').at((4,0)))
img += (slice_1 := f.RoundProcess().label('Slice').at((1, -3)))
img += (slice_2 := f.RoundProcess().label('Slice').at((7,-3)))
img += (jelly := f.Process().label('Jelly').at((-4, -3)))
img += (peanut_butter := f.Process(w=4).label('Peanut Butter').at((12, -3)))
img += (sandwich := f.Process().label('Sandwich').at((4,-6)))
img += f.Arc3(th1=0, th2=-90, arrow='->').at(start.S).to(loaf.N)
img += f.Arc3(th1=0, th2=-90, arrow='->').at(start.W).to(jelly.N)
img += f.Arc3(th1=0, th2=-90, arrow='->').at(start.E).to(peanut_butter.N)
img += f.Arc3(th1=0, th2=-90, arrow='->').at(loaf.E).to(slice_2.N)
img += f.Arc3(th1=0, th2=-90, arrow='->').at(loaf.W).to(slice_1.N)
img += f.Arc3(th1=0, th2=180, arrow='->').at(jelly.E).to(slice_1.W)
img += f.Arc3(th1=0, th2=180, arrow='->').at(peanut_butter.W).to(slice_2.E)
img += f.Arc3(th1=90, th2=180, arrow='->').at(slice_2.S).to(sandwich.E)
img += f.Arc3(th1=90, th2=180, arrow='->').at(slice_1.S).to(sandwich.W)
img += (eat := f.Process().label('Eat').at((4, -10)))
img += f.Arc3(th1=90, th2=-90, arrow='->').at(sandwich.S).to(eat.N)
img += (milk := f.Ellipse().label('Milk').at((10,-10)))
img += f.Arc3(th1=45, th2=180, arrow='<->').at(eat.E).to(milk.W)
img += (end := f.StateEnd(r=1.5).label('No Hunger').at((4, -14)))
img += f.Arc3(th1=90, th2=-90, arrow='->').at(eat.S).to(end.N)

img.save('programming_basics_img01.svg')
os.system("inkscape -f programming_basics_img01.svg -A programming_basics_img01.pdf 2> /dev/null");
img.draw()
```

```python
img = schem.Drawing()

img += (sandwich := f.Start().label('Sandwich').at((4,2.5)))
img += (bite := f.Process().label('Take Bite').at((4,0)))
img += (dry := f.Decision().label('Is mouth dry?').at((4,-3)))
img += (milk := f.Process().label('Drink Milk').at((10,-3)))
img += (more := f.Decision(w=6,h=3).label('Is there sandwich left?').at((4,-7)))
img += (end := f.StateEnd(r=1).label('Done').at((4, -12)))

img += f.Arc3(th1=0, th2=-90, arrow='->').at(sandwich.S).to(bite.N)
img += f.Arc3(th1=0, th2=-90, arrow='->').at(bite.S).to(dry.N)
img += f.Arc3(th1=0, th2=-90, arrow='->').label('Yes', ofst=(-0.4,0.2)).at(dry.E).to(milk.W)
img += f.Arc3(th1=0, th2=-90, arrow='->').label('No', ofst=(-0.25,0)).at(dry.S).to(more.N)
img += f.Arc3(th1=90, th2=180, arrow='->').at(milk.S).to(more.E)
img += f.Arc3(th1=0, th2=-90, arrow='-').label('Yes', ofst=(0.75,-1.0)).at(more.W).to((-1,-5))
img += f.Arc3(th1=90, th2=180, arrow='-').at((-1,-5)).to((1,-1))
img += f.Arc3(th1=90, th2=180, arrow='->').at((1,-1)).to(bite.W)

img += f.Arc3(th1=0, th2=-90, arrow='->').label('No',ofst=(-0.25,0)).at(more.S).to(end.N)

img.save('programming_basics_img02.svg')
os.system("inkscape -f programming_basics_img02.svg -A programming_basics_img02.pdf 2> /dev/null");
img.draw()
```

```python
img = schem.Drawing()

img += (hl := f.Process(w=18, h=9).label('High-level\nProgramming\nLanguage',ofst=(6.75,0)).at((-4,-3)))
img += (interpreted := f.Process(w=13, fill='#EEEEEE').label('Interpreted',ofst=(-5,0)).at((-3.5,0)))
img += (compiled := f.Process(w=13, h=5, fill='#EEEEEE').label('Assembled',ofst=(-5,0)).at((-3.5,-4.5)))
img += (hl_php := f.Process(h=1, fill='#FFFFFF').label('PHP').at((0,0)))
img += (hl_js := f.Process(h=1, fill='#FFFFFF').label('Javascript').at((3,0)))
img += (hl_py := f.Process(h=1, fill='#FFFFFF').label('Python').at((6,0)))

img += (hl_c := f.Process(h=1, fill='#FFFFFF').label('C').at((0,-3)))
img += (hl_cpp := f.Process(h=1, fill='#FFFFFF').label('C++').at((3,-3)))

img += (ll_as := f.Process(h=1, fill='#FFFFFF').label('Assembly').at((0,-6)))

img += (hl := f.Process(w=18, h=3).label('Low-level\nProgramming\nLanguage',ofst=(6.75,0)).at((-4,-10)))
img += (binary := f.Process(w=13, fill='#EEEEEE').label('Binary',ofst=(-5,0)).at((-3.5,-10)))
img += (mc := f.Process(h=1, w=4, fill='#FFFFFF').label('Machine Code').at((0,-10)))

img.save('programming_basics_img03.svg')
os.system("inkscape -f programming_basics_img03.svg -A programming_basics_img03.pdf 2> /dev/null");
img.draw()
```
