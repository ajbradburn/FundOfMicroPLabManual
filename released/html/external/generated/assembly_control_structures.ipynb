```python
arrow_len = 0.75

img = schem.Drawing()

# Register r0
img += (start := f.Start(w=2.5, h=1).label('Start').at((0,0)))
img += f.Arrow().length(arrow_len)
img += (load := f.Subroutine(w=4).label('Load Temp\nReading'))
img += f.Arrow().length(arrow_len)
img += (cmp_high := f.Decision(w=4, h=3).label('Is Temp &gt; 150?'))
img += f.Arrow().length(2).at(cmp_high.E).label('Yes').right()
img += (on_ac_already := f.Decision().label('AC on?'))
img += f.Arrow().length(2).label('No')
img += (on_ac := f.Process().label('Turn on\nAC'))
img += f.Arrow().length(5).at(cmp_high.S).label('No', loc='right', ofst=(0.5, -0.5)).down()
img += (cmp_low := f.Decision(w=4, h=3).label('Is Temp &lt; 100?'))
img += f.Arrow().length(2).at(cmp_low.E).label('Yes').right()
img += (on_heater_already := f.Decision().label('Heater on?'))
img += f.Arrow().length(2).label('No')
img += (on_heater := f.Process().label('Turn on\nHeater'))
img += f.Arrow().length(2).at(cmp_low.S).label('No', loc='right', ofst=(0.5, -0.5)).down()
img += (all_off := f.Process().label('All off'))

img += (end := f.Start().label('Stop').at((all_off.center.x + 16, all_off.center.y + 0.6)))

img += f.Arrow().length(2).at(on_ac_already.S).label('Yes')
img += (error := f.Process().label('Error'))
img += f.Arrow().length(2).at(on_heater_already.N).label('Yes').up()



img += f.Arc3(th1=0, th2=180, arrow='-').at(error.W).to((cmp_low.W.x, error.W.y))
img += f.Arc3(th1=0, th2=-90, arrow='-').at((cmp_low.W.x, error.W.y)).to((all_off.W.x - 2, error.W.y - 2))
img += f.Arc3(th1=90, th2=-90, arrow='-').at((all_off.W.x - 2, error.W.y - 2)).to(((all_off.W.x - 2.0), (all_off.W.y + 2)))
img += f.Arc3(th1=90, th2=180, arrow='->').at(((all_off.W.x - 2.0), (all_off.W.y + 2))).to(all_off.W)

img += f.Arc3(th1=0, th2=180, arrow='->').at(all_off.E).to(end.W)

img += f.Arc3(th1=0, th2=-90, arrow='-').at(on_ac.E).to((end.N.x, on_ac.E.y - (end.N.x - on_ac.E.x)))
img += f.Arc3(th1=0, th2=-90, arrow='-').at((end.N.x, on_ac.E.y - (end.N.x - on_ac.E.x))).to((end.N.x, on_heater.E.y))
img += f.Arc3(th1=0, th2=-90, arrow='-').at(on_heater.E).to((end.N.x, on_heater.E.y))
img += f.Dot()
img += f.Arc3(th1=0, th2=-90, arrow='->').at((end.N.x, on_heater.E.y)).to(end.N)


img.save('assembly_control_structures_img04.svg')
os.system("inkscape -f assembly_control_structures_img04.svg -A assembly_control_structures_img04.pdf 2> /dev/null");
img.draw()
```
